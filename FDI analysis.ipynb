```python
df.set_index('Sector', inplace=True)
## Reset the index to drop the default integer index
#df.reset_index(drop=True, inplace=True)
```

```python
statistics= {
    'Mean': df.mean(axis=1),
    'Median': df.median(axis=1),
    'Min': df.min(axis=1),
    'Max': df.max(axis=1),
    'Sd': df.std(axis=1)
}
```

```python
stat_df= pd.DataFrame(statistics)
stat_df
```

Based on these statistics, you can infer the following:

Sectors with higher mean and median FDI values may be considered more stable or lucrative investment opportunities.
Sectors with a wide range between minimum and maximum FDI values may have experienced fluctuations in investment inflows over the years.
Sectors with a high standard deviation may be more prone to economic or market fluctuations, posing higher risks for investors.
Sectors with lower mean and median FDI values may require targeted policies or interventions to attract more investment and stimulate growth.

```python
import matplotlib.pyplot as plt
```

```python
overall_mean = stat_df['Mean'].mean()
overall_median = stat_df['Median'].median()

stable_sectors = stat_df[(stat_df['Mean'] > overall_mean) & (stat_df['Median'] > overall_median)].index.tolist()

print("Stable Sectors with Higher Mean and Median FDI Values:")
for sector in stable_sectors:
    print(sector)
```

```python
# Extract mean and median values of stable sectors
stable_mean_values = stat_df.loc[stable_sectors, 'Mean'].values
stable_median_values = stat_df.loc[stable_sectors, 'Median'].values

# Plotting
plt.figure(figsize=(10, 6))

# Plot overall mean and median
plt.bar(['Overall Mean', 'Overall Median'], [overall_mean, overall_median], color='gray', label='Overall')

# Plot mean and median of stable sectors
plt.bar(['Stable Mean', 'Stable Median'], [stable_mean_values.mean(), stable_median_values.mean()], color='blue', label='Stable Sectors')

plt.title('Comparison of Mean and Median for Stable Sectors vs Overall')
plt.ylabel('FDI')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculate the range of FDI values for each sector
stat_df['Range'] = stat_df[['Min', 'Max']].apply(lambda x: x['Max'] - x['Min'], axis=1)
stat_df.head()

```

```python
# Sort the sectors based on the range in descending order
sectors_with_wide_range = stat_df.sort_values(by='Range', ascending=False)

# Display the sectors with the widest range
print("Sectors with a Wide Range between Minimum and Maximum FDI Values:")
print(sectors_with_wide_range[['Min', 'Max', 'Range']].head())
```

```python
# Calculate the overall minimum and maximum FDI values across all sectors
overall_min = stat_df['Min'].min()
overall_max = stat_df['Max'].max()

# Calculate the overall range
overall_range = overall_max - overall_min

print("Overall Range of FDI Values across All Sectors:", overall_range)
```

assumption : if the sector range exceeds 10% of my overall range then it is considered as a sector which has experienced fluctuations

```python
# Calculate the threshold for identifying sectors with wide fluctuations
threshold = 0.1 * overall_range
print("Threshold:", threshold)

# Filter sectors where the range exceeds the threshold
fluctuating_sectors = sectors_with_wide_range[sectors_with_wide_range['Range'] > threshold]

# Display the fluctuating sectors
print("Sectors with Fluctuations Exceeding 10% of Overall Range:")
print(fluctuating_sectors[['Min', 'Max', 'Range']])
```

```python
# Find the intersection of the two lists
common_sectors = set(fluctuating_sectors) & set(stable_sectors)

# Display the intersection (common sectors)
print("Intersection of Fluctuating Sectors and Stable Sectors:")
print(len(common_sectors))
```

so, we can say that the stable_sectors do not have a wide fluctuations

```python
overall_dev= stat_df['Sd'].std()
print(overall_dev)
```

```python
dev_df= stat_df[stat_df['Sd'] > overall_dev]
dev_df
```

These sectors have a standard deviation higher than the average standard deviation of the total FDI.
So, these sectors experience higher fluctuations. We must find out which number of sectors in our stable_sector df belong to this category.

```python
# Find the intersection of the two lists
common_sectors = set(dev_df) & set(stable_sectors)

# Display the intersection (common sectors)
print("Intersection of Fluctuating Sectors and Stable Sectors:")
print(len(common_sectors))
```

We see that no number of sectors in our stable_sector df belong to this category.

```python
# Initialize a dictionary to store the total FDI for each stable sector
total_fdi_per_sector = {}

# Iterate over each stable sector
for sector in stable_sectors:
    # Filter the DataFrame for the current sector
    sector_df = df.loc[sector]
    
    # Calculate the total FDI for the current sector
    total_fdi = sector_df.sum()  # Sum across all years
    
    # Store the total FDI for the current sector in the dictionary
    total_fdi_per_sector[sector] = total_fdi

# Display the total FDI for each stable sector
for sector, total_fdi in total_fdi_per_sector.items():
    print(f"Total FDI for {sector}: {total_fdi}")
```

```python
# Convert the dictionary total_fdi_per_sector to lists
sectors = list(total_fdi_per_sector.keys())
total_fdi = list(total_fdi_per_sector.values())

# Plotting
plt.figure(figsize=(14, 8))
plt.bar(sectors, total_fdi, color='skyblue')
plt.xlabel('Sector')
plt.ylabel('Total FDI (in million $)')
plt.title('Total FDI for Stable Sectors')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
